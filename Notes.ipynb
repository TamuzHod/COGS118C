# Event Related Analysis
### To Overcome Noise make the foloing assumptions:
* Same stimulus generates the same "true" response 
* Noise is statisticly independent
* Noise does not change over time

Time zero is knowen as triger event

### Event-Related Potential (ERP) or Evoked Potential
We can avrage a number of trials to "avrage out" the noise when the noise is:
* Independent of any experimantal variable 
* Stationary statistical measures (mean, SD, etc.) do not change over time
* Ergodic avraging over "ensembles" (trials) is equivalent to avreging over time
    * Fliping a coin 10 times + Fliping 10 coins 1 time when result is avraged

#### Procedure for (ERA)
* Record raw data
* preprocess data
* Find event triggers corresponding to stimulus presentation
* For each trigger, cut a window of data around it (epoching)
* Subtract (pre-stimulus) baseline to remove slow drifts
* Avrage over groups/conditions

#### Effect of Averging
Becuse noise is independent:
* Avarging over N trials maintains signal rms while decreasing noise rms
    * rms_avrage_signal ~ rms_single_trial_signal
    * rms_avrage_noise ~ rms_single_trial_noise / sqrt (N)
